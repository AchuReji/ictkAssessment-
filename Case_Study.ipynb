```python
import pandas as pd
```

```python
data=pd.read_csv('auto-mpg (1).csv')
```

```python
data.head()
```

Basic Array Operations

```python
import numpy as np


mpg = data['mpg'].to_numpy()

mean_mpg = np.mean(mpg)
median_mpg = np.median(mpg)
std_mpg = np.std(mpg)
cars_above_25 = np.sum(mpg > 25)
```

```python
print("Mean MPG:", mean_mpg)
print("Median MPG:", median_mpg)
print("Standard Deviation of MPG:", std_mpg)
print("Number of cars with MPG > 25:", cars_above_25)
```

2. Filtering

```python
filtered = data[data['cylinders'] > 6]
car_names = filtered['car name'].tolist()
print("Cars with more than 6 cylinders:", car_names)
```

```python
print("Cars with more than 6 cylinders:", car_names)
```

Statistical Analysis

```python
weight = data['weight'].to_numpy()
percentiles = np.percentile(weight, [25, 50, 75])
print("Weight Percentiles (25th, 50th, 75th):", percentiles)
```

Array Manipulation

```python
acceleration = data['acceleration'].to_numpy()
normalized_acc = (acceleration - np.min(acceleration)) / (np.max(acceleration) - np.min(acceleration))
print("First 5 normalized acceleration values:", normalized_acc[:5])
```

Broadcasting

```python
horsepower = data['horsepower'].replace('?', np.nan).astype(float)

# Handle missing values
horsepower.fillna(horsepower.mean(), inplace=True)

# Increase by 10%
new_hp = horsepower.to_numpy() * 1.10
print("First 5 updated horsepower values:", new_hp[:5])
```

Boolean Indexing

```python
# Origin 2 represents Europe
europe_cars = data[data['origin'] == 2]
avg_displacement = np.mean(europe_cars['displacement'])
print("Average displacement of European cars:", avg_displacement)
```

Matrix Operations

```python
matrix = data[['mpg', 'horsepower', 'weight']].replace('?', np.nan).astype(float).to_numpy()
vector = np.array([1, 0.5, -0.2])
dot_product = np.dot(matrix, vector)
print("First 5 results of dot product:", dot_product[:5])
```

Sorting

```python
sorted_data = data.sort_values(by='model year', ascending=False)
top_5 = sorted_data['car name'].head(5).tolist()
print("Top 5 cars by model year:", top_5)
```

Correlation

```python
correlation = np.corrcoef(data['mpg'], data['weight'])[0, 1]
print("Correlation between MPG and Weight:", correlation)
```

Conditional Aggregates

```python
mean_mpg_by_cyl = data.groupby('cylinders')['mpg'].mean()
print("Mean MPG by Cylinders:")
print(mean_mpg_by_cyl)
```

Pandas Problems

1. Basic Exploration

```python
# First 10 rows
print("First 10 Rows:\n", data.head(10))

# Shape of dataset
print("\nTotal Rows and Columns:", data.shape)

# Summary statistics
print("\nSummary Statistics:\n", data.describe())
```

2. Filtering and Indexing

```python
filtered_data = data[(data['model year'] == 75) & (data['weight'] < 3000)]
filtered_data = filtered_data[['car name', 'weight', 'mpg']]
print("\nCars from 1975 with weight < 3000:\n", filtered_data)
```

Handling Missing Data

```python
# Check missing values
print("\nMissing Values Before Handling:\n", data.isna().sum())

# Replace missing horsepower with median
data['horsepower'] = data['horsepower'].replace('?', np.nan).astype(float)
data['horsepower'].fillna(data['horsepower'].median(), inplace=True)

print("\nMissing Values After Handling:\n", data.isna().sum())
```

Data Transformation

```python
data['power_to_weight_ratio'] = data['horsepower'] / data['weight']
print("\nAdded 'power_to_weight_ratio' column.")
data.head()
```

Group By

```python
avg_mpg_by_origin = data.groupby('origin')['mpg'].mean()
print("\nAverage MPG by Origin:\n", avg_mpg_by_origin)
```

Sorting

```python
top_10_mpg = data.sort_values(by='mpg', ascending=False).head(10)
print("\nTop 10 Cars by MPG:\n", top_10_mpg[['car name', 'mpg']])
```

Apply Function

```python
# Define the function
def performance_score(row):
    return (row['mpg'] * row['acceleration']) / row['weight']

# Apply to dataframe
data['performance_score'] = data.apply(performance_score, axis=1)
print("\nAdded 'performance_score' column.\n")
data[['car name', 'performance_score']].head()
```

Visualization Preparation

```python
summary = data.groupby('model year')[['mpg', 'weight', 'horsepower']].mean()
print("\nSummary DataFrame (Average mpg, weight, horsepower per model year):\n", summary.head())
```

Exporting Data

```python
high_mpg = data[data['mpg'] > 30][['mpg', 'cylinders', 'horsepower', 'weight']]
high_mpg.to_csv('high_mpg_cars.csv', index=False)
print("\nExported 'high_mpg_cars.csv' for cars with mpg > 30")
```

Finding Anomalies (IQR Method)

```python
Q1 = data['mpg'].quantile(0.25)
Q3 = data['mpg'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = data[(data['mpg'] < lower_bound) | (data['mpg'] > upper_bound)]
print("\nIQR for MPG:", IQR)
print("\nOutliers in MPG:\n", outliers[['car name', 'mpg', 'model year']])
```
