```python
orders = [
    {"item": "Laptop", "quantity": 2, "price": 50000},
    {"item": "Mouse", "quantity": 10, "price": 500},
    {"item": "Keyboard", "quantity": 5, "price": 1500},
    {"item": "Monitor", "quantity": 1, "price": 10000},
    {"item": "Headphones", "quantity": 3, "price": 2000},
    {"item": "Laptop", "quantity": 1, "price": 50000},
    {"item": "Mouse", "quantity": 5, "price": 500},
    {"item": "Keyboard", "quantity": 2, "price": 1500},
]
```

 Calculate the total revenue for the store

```python
total_revenue = sum(order["quantity"] * order["price"] for order in orders)
print(f"2.1 Total Revenue: ₹{total_revenue}")
```

Identify all orders where the total order value

```python
high_value_orders = [order for order in orders if order["quantity"] * order["price"] > 2000]
print("2.2 Orders with total value > ₹2000:")
for order in high_value_orders:
    print(order)
```

Find the item with the highest total quantity sold across all orders.

```python
from collections import defaultdict
quantity_per_item = defaultdict(int)
for order in orders:
    quantity_per_item[order["item"]] += order["quantity"]
max_item = max(quantity_per_item, key=quantity_per_item.get)
max_quantity = quantity_per_item[max_item]
print(f"2.3 Item with highest total quantity sold: {max_item} (Quantity: {max_quantity})")
```

3.1 question


```python
orders = [
    {'item': 'apple', 'price_per_item': 1.00, 'quantity': 5},
    {'item': 'banana', 'price_per_item': 0.50, 'quantity': 10},
    {'item': 'orange', 'price_per_item': 0.75, 'quantity': 8}
]
```

```python

def apply_discount(orders, discount_rate):

    discounted_orders = []
    for order in orders:

        new_order = order.copy()
        new_order["price_per_item"] = round(order["price_per_item"] * (1 - discount_rate), 2)
        discounted_orders.append(new_order)
    return discounted_orders
#10% discount
discounted_list = apply_discount(orders, 0.10)

print("Original Orders:")
for o in orders:
    print(o)

print("\nDiscounted Orders:")
for d in discounted_list:
    print(d)
```

4.1&2 question

```python

orders = [
    {"order_id": 1, "customer": "sudhi", "item": "BMW", "quantity": 2, "price_per_item": 16653599.00},
    {"order_id": 2, "customer": "Neha ", "item": "AUDI", "quantity": 1, "price_per_item": 14653599.50},
    {"order_id": 3, "customer": "rahul", "item": "LANDOVER", "quantity": 1, "price_per_item":17653599.00},
    {"order_id": 4, "customer": "roshni", "item": "Porsche 911", "quantity": 3, "price_per_item": 18653599.00},
]


class OrderIterator:
    def __init__(self, orders):
        self.orders = orders
        self.index = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.index < len(self.orders):
            order = self.orders[self.index]
            self.index += 1
            return order
        else:
            raise StopIteration


print("Iterating through orders using custom iterator:\n")
order_iter = OrderIterator(orders)

for order in order_iter:
    print(order)
```

```python

grouped_orders = {
    customer: [order for order in orders if order["customer"] == customer]
    for customer in {order["customer"] for order in orders}
}

print("\nOrders Grouped by Customer:\n")
for customer, order_list in grouped_orders.items():
    print(f"{customer}:")
    for o in order_list:
        print("   ", o)
```
